```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Read the dataset
df = pd.read_csv('car_insurance_premium_dataset.csv')

# Calculate correlation matrix with all numeric features
correlation_matrix = df.corr()

# Create a heatmap of correlations
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Insurance Premium Factors')
plt.tight_layout()
plt.show()

# Calculate non-linear correlations (Spearman)
spearman_corr = df.corr(method='spearman')
print('\nSpearman Correlation with Insurance Premium:')
print(spearman_corr['Insurance Premium ($)'].sort_values(ascending=False))

# Calculate interaction effects
df['Age_Experience_Interaction'] = df['Driver Age'] * df['Driver Experience']
df['Risk_Score'] = df['Previous Accidents'] * (1 / (df['Driver Experience'] + 1))

# Analyze the relationship between Risk Score and Premium
plt.figure(figsize=(10, 6))
plt.scatter(df['Risk_Score'], df['Insurance Premium ($)'])
plt.xlabel('Risk Score (Accidents/Experience)')
plt.ylabel('Insurance Premium ($)')
plt.title('Risk Score vs Insurance Premium')
plt.show()

# Calculate summary statistics for different risk groups
df['Risk_Category'] = pd.qcut(df['Risk_Score'], q=3, labels=['Low', 'Medium', 'High'])
print('\nPremium Statistics by Risk Category:')
print(df.groupby('Risk_Category')['Insurance Premium ($)'].describe())

# Perform PCA to identify hidden patterns
scaler = StandardScaler()
features = ['Driver Age', 'Driver Experience', 'Previous Accidents', 'Annual Mileage (x1000 km)', 'Car Age']
X_scaled = scaler.fit_transform(df[features])
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Print explained variance ratio
print('\nPCA Explained Variance Ratio:')
print(pd.DataFrame(
    {'Explained Variance Ratio': pca.explained_variance_ratio_,
     'Cumulative Variance Ratio': np.cumsum(pca.explained_variance_ratio_)},
    index=[f'PC{i+1}' for i in range(len(features))]
))
```

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('car_insurance_premium_dataset.csv')

# Age Groups Analysis
df['Age_Group'] = pd.cut(df['Driver Age'], bins=[0, 25, 35, 45, 55, 100], labels=['18-25', '26-35', '36-45', '46-55', '55+'])
plt.figure(figsize=(12, 6))
sns.boxplot(x='Age_Group', y='Insurance Premium ($)', data=df)
plt.title('Premium Distribution Across Age Groups')
plt.show()

# Experience vs Premium with Accident Count
plt.figure(figsize=(10, 6))
for accidents in df['Previous Accidents'].unique():
    subset = df[df['Previous Accidents'] == accidents]
    plt.scatter(subset['Driver Experience'], 
               subset['Insurance Premium ($)'], 
               label=f'Accidents: {accidents}',
               alpha=0.6)
plt.xlabel('Driver Experience (Years)')
plt.ylabel('Insurance Premium ($)')
plt.title('Experience vs Premium by Accident Count')
plt.legend()
plt.show()

# Mileage Impact Analysis
plt.figure(figsize=(10, 6))
sns.regplot(x='Annual Mileage (x1000 km)', 
           y='Insurance Premium ($)', 
           data=df,
           scatter_kws={'alpha':0.5})
plt.title('Mileage vs Premium with Regression Line')
plt.show()

# Calculate complex metrics
print('\nKey Statistics by Age Group:')
age_stats = df.groupby('Age_Group').agg({
    'Insurance Premium ($)': ['mean', 'std'],
    'Previous Accidents': 'mean',
    'Annual Mileage (x1000 km)': 'mean'
}).round(2)
print(age_stats)

# Premium variation by car age and mileage
df['Mileage_Category'] = pd.qcut(df['Annual Mileage (x1000 km)'], q=3, labels=['Low', 'Medium', 'High'])
print('\nPremium Statistics by Car Age and Mileage:')
pivot_table = pd.pivot_table(df, 
                           values='Insurance Premium ($)',
                           index='Car Age',
                           columns='Mileage_Category',
                           aggfunc='mean').round(2)
print(pivot_table.head())
```
